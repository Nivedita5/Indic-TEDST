```python
import os
import requests
from requests.exceptions import ReadTimeout
import time
import nltk
from nltk.tokenize import sent_tokenize
from contextlib import redirect_stdout

nltk.download('punkt')

test_alignments = []

output_dir_eng = '/content/drive/MyDrive/hi_en'
output_dir_hindi = '/content/drive/MyDrive/hi'
os.makedirs(output_dir_eng, exist_ok=True)
os.makedirs(output_dir_hindi, exist_ok=True)

retry_delay = 10
max_retries = 5

for file in os.listdir(src_dir):
    src_file = os.path.join(src_dir, file).replace("\\", "/")
    tgt_file = os.path.join(tgt_dir, file).replace("\\", "/")

    if os.path.isfile(src_file) and os.path.isfile(tgt_file) and os.path.getsize(src_file) > 0 and os.path.getsize(tgt_file) > 0:
        src = open(src_file, 'rt', encoding='utf-8', errors='replace').read()
        tgt = open(tgt_file, 'rt', encoding='utf-8', errors='replace').read()

        retries = 0
        while retries < max_retries:
            try:
                print("Start aligning {} to {}".format(src_file, tgt_file))
                aligner = Bertalign(src, tgt, is_split=True)

                aligner.align_sents()
                if aligner.result is None:
                    print("Alignment failed for {}. Skipping file.".format(src_file))
                    break

                test_alignments.append(aligner.result)

                # Output English and Hindi alignments to separate files
                output_file_eng = os.path.join(output_dir_eng, file.replace('.txt', '_eng_alignment.txt'))
                output_file_hindi = os.path.join(output_dir_hindi, file.replace('.txt', '_hindi_alignment.txt'))

                with open(output_file_eng, 'w', encoding='utf-8') as output_eng:
                    with open(output_file_hindi, 'w', encoding='utf-8') as output_hindi:
                        with redirect_stdout(output_eng):
                            aligner.print_sents()
                        with redirect_stdout(output_hindi):
                            # Split Hindi text into sentences using NLTK
                            hindi_sentences = sent_tokenize(tgt)
                            for sentence in hindi_sentences:
                                print(sentence)

                break

            except ReadTimeout:
                retries += 1
                print("ReadTimeout occurred for {}. Retrying in {} seconds...".format(src_file, retry_delay))
                time.sleep(retry_delay)

            except KeyError as e:
                print("KeyError occurred: {}".format(e))
                print("Skipping file: {}".format(src_file))
                break

    else:
        print("Source or target file is missing or empty for {}. Skipping file.".format(src_file))
```
